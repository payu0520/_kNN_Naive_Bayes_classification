```python
# Importing libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
# Importing and reading dataset

ds = pd.read_csv("Social_Network_ads.csv")
ds
```

```python
ds.columns
```

```python
ds.duplicated().sum()
```

```python
ds.drop_duplicates(inplace = True)
```

```python
ds.isnull().sum()
```

```python
# Splitting data as Independent and Dependent variables

X = ds[['Age', 'EstimatedSalary']]
y = ds[['Purchased']]
```

```python
# Feature scaling

from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X = sc.fit_transform(X)
```

```python
# Dividing dataset into Training and Testing data

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

```python
# Classification model fitting

from sklearn.neighbors import KNeighborsClassifier
cls = KNeighborsClassifier(n_neighbors = 5, metric = 'minkowski', p=2)
cls.fit(X_train, y_train)
```

```python
# predict for my input

cls.predict(sc.transform([[40,97000]]))
```

```python
# Concatenate the prediction and original output

y_pred = cls.predict(X_test)
y_pred = pd.DataFrame(data = y_pred.reshape(100,1), columns = ["Prediction"])
```

```python
y_test
```

```python
result = pd.DataFrame(np.concatenate((y_test, y_pred), axis = 1), columns = ["Purchased","Prediction"])
result
```

```python
# Cheaking accuracy using confusion matrix

from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(result["Purchased"], result["Prediction"])
cm
```

```python
score = accuracy_score(result["Purchased"], result["Prediction"])       # we get 93% accuracy in prediction
score
```

```python
# User Input

age = float(input("Enter Age: "))
salary = float(input("Enter Estimated Salary: "))

myinput = pd.DataFrame(sc.transform([[age, salary]]), columns=['Age', 'EstimatedSalary'])

result = cls.predict(myinput)

if result[0] == 1:
    print("Prediction: User will purchase plan")
else:
    print("Prediction: User will not purchase plan")
```
